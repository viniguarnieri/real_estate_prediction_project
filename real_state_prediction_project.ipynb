```python
# Libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import joblib
%matplotlib inline


def show_sample(data):
    print('                                             Initial Visualization                                                ')
    print('------------------------------------------------------------------------------------------------------------------')
    print('                                               Sample of 10 rows                                                  ')
    display(data.sample(10))
    print('==================================================================================================================')
    return None
    
def show_data_info(data):
    print('                                                  Data infos                                                      ')
    print(data.info())
    print('==================================================================================================================')
    return None


def plot_price_per_unit_statistics(data):
    
    print('                                                     EDA                                                          ')
    print('------------------------------------------------------------------------------------------------------------------')
    
    print('                                                price_per_unit                                                    ')
    # Get the label column
    label = data['price_per_unit']

    # Get statistics
    label_min = label.min()
    label_max = label.max()
    label_mean = label.mean()
    label_median = label.median()
    label_mode = label.mode()[0]

    print('Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(label_min, 
                                                                                             label_max, 
                                                                                             label_mean, 
                                                                                             label_median, 
                                                                                             label_mode))
    # Create subplots
    fig, ax = plt.subplots(2,1, figsize=(7,9))

    # Plot histogram
    ax[0].hist(label)

    # Add lines for the min, max, mean, median, and mode
    ax[0].axvline(label_min, color='gray', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_max, color='gray', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_mean, color='cyan', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_median, color='magenta', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_mode, color='yellow', linestyle='dashed', linewidth=2)

    # Plot boxplot
    ax[1].boxplot(label, vert=False)

    # Titles
    for i in range(0,2):
        ax[i].set_title('Data Distribution')
        ax[i].set_xlabel('Price')
        ax[i].set_ylabel('Frequency')
    
    # Show
    plt.show()
    print('==================================================================================================================')
    return data

def remove_outliers_and_plot_again(data):
    
    print('                                       price_per_unit (without outliers)                                          ')
    # Remove outliers
    data = data[data['price_per_unit']<70]

    # Get the label column
    label = data['price_per_unit']

    # Get statistics
    label_min = label.min()
    label_max = label.max()
    label_mean = label.mean()
    label_median = label.median()
    label_mode = label.mode()[0]

    print('Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(label_min, 
                                                                                             label_max, 
                                                                                             label_mean, 
                                                                                             label_median, 
                                                                                             label_mode))
    # Create subplots
    fig, ax = plt.subplots(2,1, figsize=(7,9))

    # Plot histogram
    ax[0].hist(label)

    # Add lines for the min, max, mean, median, and mode
    ax[0].axvline(label_min, color='gray', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_max, color='gray', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_mean, color='cyan', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_median, color='magenta', linestyle='dashed', linewidth=2)
    ax[0].axvline(label_mode, color='yellow', linestyle='dashed', linewidth=2)

    # Plot boxplot
    ax[1].boxplot(label, vert=False)

    # Titles
    for i in range(0,2):
        ax[i].set_title('Data Distribution')
        ax[i].set_xlabel('Price')
        ax[i].set_ylabel('Frequency')
    
    # Show
    plt.show()
    print('==================================================================================================================')
    return data
    
def plot_numeric_features(data):
    
    # Remove outliers
    data = data[data['price_per_unit']<70]
    
    print('                                              Numeric Features                                                    ')
    # Numeric features (There is not categorical features so the numeric features is going to be the data without 'price_per_unit')
    numeric_features = data.drop('price_per_unit', axis=1)
    
    # Get label
    label = data['price_per_unit']
    
    # Plot a histogram for each numeric feature
    for col in numeric_features:
        fig = plt.figure(figsize=(8,5))
        ax = fig.gca()
        feature = data[col]
        correlation = feature.corr(label)
        plt.scatter(x=feature, y=label)
        ax.axvline(feature.mean(), color='magenta', linestyle='dashed', linewidth=2)
        ax.axvline(feature.median(), color='cyan', linestyle='dashed', linewidth=2)
        plt.xlabel(col)
        plt.ylabel('Correlations')
        ax.set_title('price_per_unit vs ' + col + ' - correlation: ' + str(correlation))
    
    plt.show()
    print('==================================================================================================================')
    return None 

def plot_boxplots(data):
    
    # Remove outliers
    data = data[data['price_per_unit']<70]
    
    print('                                                  Box Plots                                                       ')
    label = data[['local_convenience_stores', 'transaction_date']]
    
    # plot a boxplot for the label by each categorical feature
    for col in label:
        fig = plt.figure(figsize=(9, 6))
        ax = fig.gca()
        data.boxplot(column = 'price_per_unit', by = col, ax = ax)
        ax.set_title('Label by ' + col)
        plt.xticks(rotation=45)
        
    plt.show()
    print('==================================================================================================================')
    return None

def machine_learning(data):
    
    # Remove outliers
    data = data[data['price_per_unit']<70]
    
    # Pearson Correlation
    print('                                            Pearson Correlation                                                   ')
    plt.figure(figsize=(15,10))
    cor = data.corr()
    sns.heatmap(cor, annot=True, cmap='crest')
    plt.show()
    print('==================================================================================================================')

    # Correlation with target variable price
    print('                                             Relevant Features                                                    ')
    cor_target = abs(cor['price_per_unit'])
    relevant_features = cor_target
    print(relevant_features)
    print('==================================================================================================================')
    
    # 'transaction_date' do not have a good correlation so its gonna be dropped.
    data = data.drop('transaction_date', axis=1)
    
    # Input features
    X = data.drop('price_per_unit', axis=1)

    # Target 
    y = data['price_per_unit']

    # Split data 70%-30% into training set and test set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
    print('                                      Shape of training and test sets                                             ')
    print('Training set: %d rows \nTest set: %d rows' % (X_train.shape[0], X_test.shape[0]))
    
    # Linear Regression
    # Fit training data
    print('==================================================================================================================')
    print('                                             Linear Regression                                                    ')
    lr = LinearRegression().fit(X_train, y_train)

    # Predict the y_test
    y_pred = lr.predict(X_test)


    # Evaluate the model on the testing data
    lr_mse = mean_squared_error(y_test, y_pred)
    lr_rmse = np.sqrt(lr_mse)
    lr_r2 = r2_score(y_test, y_pred)

    print("Mean squared error:", lr_mse)
    print("RMSE :", lr_rmse)
    print("R-squared:", lr_r2)

    # Decision Tree Regressor
    # Fit
    print('==================================================================================================================')
    print('                                               Decision Tree                                                      ')
    dt = DecisionTreeRegressor().fit(X_train, y_train)

    # Predict
    y_pred = dt.predict(X_test)

    # Evaluate the model on the testing data
    dt_mse = mean_squared_error(y_test, y_pred)
    dt_rmse = np.sqrt(dt_mse)
    dt_r2 = r2_score(y_test, y_pred)

    print("Mean squared error:", dt_mse)
    print("RMSE :", dt_rmse)
    print("R-squared:", dt_r2)

    # Random Forest Regressor
    # Fit
    print('==================================================================================================================')
    print('                                               Random Forest                                                      ')
    rf = RandomForestRegressor().fit(X_train, y_train)

    # Predict
    y_pred = rf.predict(X_test)

    # Evaluate the model on the testing data
    rf_mse = mean_squared_error(y_test, y_pred)
    rf_rmse = np.sqrt(rf_mse)
    rf_r2 = r2_score(y_test, y_pred)

    print("Mean squared error:", rf_mse)
    print("RMSE :", rf_rmse)
    print("R-squared:", rf_r2)

    # Gradient Boosting Regressor
    # Fit
    print('==================================================================================================================')
    print('                                             Gradient Boosting                                                    ')
    gb = GradientBoostingRegressor().fit(X_train, y_train)

    # Predict
    y_pred = gb.predict(X_test)
    
    # Evaluate the model on the testing data
    gb_mse = mean_squared_error(y_test, y_pred)
    gb_rmse = np.sqrt(gb_mse)
    gb_r2 = r2_score(y_test, y_pred)

    print("Mean squared error:", gb_mse)
    print("RMSE :", gb_rmse)
    print("R-squared:", gb_r2)
    
    # New data to test the Artificial Intelligence (AI)
    d = {'transaction_date': [2013.167, 2013.000],
    'house_age': [16.2, 13.6],
    'transit_distance': [289.3248, 4082.015],
    'local_convenience_stores': [5, 0],
    'latitude': [24.98203, 24.94155],
    'longitude': [121.54348, 121.50381]}

    new_data = pd.DataFrame(d)
    
    # Create the column 'price_per_unit' in new_data
    new_data['price_per_unit'] = 'N/A'
    
    # Setting the data that will used in AI
    X = new_data.drop(['price_per_unit', 'transaction_date'], axis=1)
    
    print('==================================================================================================================')
    print('                                           New data to test the AI                                                ')
    display(new_data)
    print('\n')
    
    # Save the Random Forest model as a pickle file
    filename = './rental-estate.pkl'
    joblib.dump(rf, filename)

    # Load the model from the file
    loaded_model = joblib.load(filename)

    # Predict the column 'price_per_unit' of the new_data
    results = loaded_model.predict(X)

    # Load the results of the predicting in a DataFrame called 'results'
    print('==================================================================================================================')
    print('                      Results of applying the trained Random Forest model in new_data                             ')
    
    results = pd.DataFrame(results).reset_index(drop=True)
    
    # Change the name of the column to 'price_per_unit'
    results.columns = ['price_per_unit']
    display(results)
    
    return None
    
def data_collect (path):
    # Load dataset
    data = pd.read_csv(path)
    
    show_sample(data)
    
    show_data_info(data)
    
    return data


def data_transform (data):

    plot_price_per_unit_statistics(data)
    
    remove_outliers_and_plot_again(data)
    
    plot_numeric_features(data)
    
    plot_boxplots(data)
    
    machine_learning(data)
    
    return data


if __name__ == '__main__':
    #ETL
    data_raw = data_collect ('real_estate.csv')
    
    data_processing = data_transform (data_raw)
```
